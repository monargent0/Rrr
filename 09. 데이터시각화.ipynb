# 학력별 점수차이

```r
data <- read.csv("../Data/education.csv", header = F)
colnames(data) <- c('성별', '점수',  '과목','학력')
head(data)
```

```r
str(data)
summary(data)
```

```r
# V1 남여구분 / V2 점수 / V3 국,영,수 / V4 초,중,고,대,원 
# factor 사용! (숫자와 문자값을 갖고 있는 타입)
# 컬럼 이름 변경(성별,점수,과목,학력)
data$성별 <- factor(data$성별, levels = 1:2, labels = c('남', '여'))
data$과목 <- factor(data$과목, levels = 1:3, labels = c('국어', '영어','수학'))
data$학력 <- factor(data$학력, levels = 1:5, labels = c('초', '중', '고','대','원'))
```

```r
str(data)
```

```r
# 막대 그래프로 점수 빈도수를 표현
# 막대그래프는 table과 친함
par(family = 'AppleGothic', bg = 'mistyrose')
barplot(table(data$점수), 
    main = '점수별 빈도수',
    xlab = '점수',
    ylab = '빈도수',
    ylim = c(0,12),
   
)
```

```r
# 성별 학력별 빈도수를 표현
table(data$성별)
table(data$학력)
table(data$성별, data$학력)
table(data$학력 , data$성별)
```

```r
# 성별, 학력의 빈도수를 막대그래프로 그리기
par(family = 'AppleGothic', bg = 'mistyrose')
barplot(table(data$성별, data$학력), 
    beside = T,
    main = '성별 학력별 빈도수',
    xlab = '학력별',
    ylab = '빈도수',
    ylim = c(0,6),
    col = c('orange','cornsilk'),
    legend.text = T

)
```

```r
sales <- c(210,110,400,550,600,130)
week <- c('월','화','수','목','금','토')
ratio <- sales/ sum(sales)*100 
label <- paste(week, "\n", ratio, "%")
```
