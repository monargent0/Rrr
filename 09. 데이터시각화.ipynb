```r
### 서울시 교통사고 데이터로 발생건수와 부상자수의 산포도 구하기
seoul <- read.csv("../Data/newSeoul2019.csv")
head(seoul)
```

```r
par(family = "AppleGothic")
plot(seoul$발생건수, 
    seoul$부상자수, 
    main= "서울시 교통사고 발생건수와 부상자수 산포도",
    xlab="발생건수", 
    ylab='부상자수',
    col = ifelse( seoul$발생건수 >= 150, 'blue','yellow')
)
```

```r
# main title : Simple Time Series
y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    main = "Simple Time Series",
    xlim = c(min(y)-1,max(y)+1),
    ylim = c(min(y)-1,max(y)+1),
    #col = rainbow(length(y))
    col = 'white'
    )
# 시간데이터는 sort하면 안된다
```

```r
# 꺽은선 그래프
y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    main = "Simple Time Series",
    xlim = c(min(y)-1,max(y)+1),
    ylim = c(min(y)-1,max(y)+1),
    col = 'white',
    type = 'line'
    )
```

```r
# 점과 꺽은선 그래프
y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    main = "Simple Time Series",
    xlim = c(min(y)-1,max(y)+1),
    ylim = c(min(y)-1,max(y)+1),
    col = 'white',
    type = 'both'
    )
```

```r
# 점이 빠진 꺽은선 그래프 
y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    main = "Simple Time Series",
    xlim = c(min(y)-1,max(y)+1),
    ylim = c(min(y)-1,max(y)+1),
    col = 'white',
    type = 'c'
    )
```

```r
# 점과 꺽은선을 중첩한 그래프
y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    main = "Simple Time Series",
    xlim = c(min(y)-1,max(y)+1),
    ylim = c(min(y)-1,max(y)+1),
    col = 'white',
    type = 'o'
    )
```

```r
# 수직선 그래프
y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    main = "Simple Time Series",
    xlim = c(min(y)-1,max(y)+1),
    ylim = c(min(y)-1,max(y)+1),
    col = 'purple',
    type = 'h'
    )
```

```r
# 왼쪽 값을 기초로한 계단 모양 그래프
y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    main = "Simple Time Series",
    xlim = c(min(y)-1,max(y)+1),
    ylim = c(min(y)-1,max(y)+1),
    col = 'purple',
    type = 's'
    )
```

```r
# 오른쪽 값을 기초로한 계단 모양 그래프
y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    main = "Simple Time Series",
    xlim = c(min(y)-1,max(y)+1),
    ylim = c(min(y)-1,max(y)+1),
    col = 'purple',
    type = 'S'
    )
```

```r
# 선 타입 
y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    main = "Simple Time Series",
    xlim = c(min(y)-1,max(y)+1),
    ylim = c(min(y)-1,max(y)+1),
    col = 'white',
    type = 'l',
    lty = 5 # linetype
    )
```

```r
# 포인트
y <- c(1,3,2,4,8,5,7,6,9)

plot(y,
    main = "Simple Time Series",
    xlim = c(min(y)-1,max(y)+1),
    ylim = c(min(y)-1,max(y)+1),
    col = 'white',
    type = 'p',
    lty = 5, # linetype
    pch = c(11, 20,9)
)
```

```r
x <- runif(100)
y <- runif(100)

# y = 0.5 이면 pch이면 1, 아니면 18
# col : y>0.5이면 ;'red', 아니면 "blue"

plot(x,
    y,
    main = "Simple Time Series",
    xlim = c(min(y)-0.1,max(y)+0.1),
    ylim = c(min(y)-0.1,max(y)+0.1),
    col = ifelse( y > 0.5 , 'mistyrose', 'skyblue'),
    pch = ifelse( y > 0.5 , 8, 18)
)
```
