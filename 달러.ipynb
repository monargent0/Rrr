```r
data <- read.csv('../AIData/진짜덩대.csv', header = TRUE, stringsAsFactors = FALSE)
data <- data[,-(2:3)]
data <- data[,-(28:53)]
data <- data[,c(1,6,2,4,7,11,12,14,18,20,24,26)]
colnames(data) <- c('날짜','금','원달러','유로달러','브렌트유','TED','미국10년국채','나스닥','상하이','코스피','VIX','미국CLI')
str(data)
```

```r
head(data)
data.S <- cbind(data[,1:2], scale(data[,-(1:2)]))
head(data.S)
data.C <- data.S[,-(1:3)]
```

```r
data.kmeans <- kmeans(data.S[,-(1:2)], centers = 3, iter.max = 10000)
data.kmeans$centers
```

```r
data.S$cluster <- as.factor(data.kmeans$cluster)
```

```r
par(bg='white')
plot(data.S$금 , col = data.S$cluster)
```
