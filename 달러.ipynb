```r
data <- read.csv('../AIData/진짜덩대.csv', header = TRUE, stringsAsFactors = FALSE)

doil <- data[,c(1:5, 9:12 )]
```

```r
head(doil)
str(doil)
```

```r
options(warn = -1)
options(scipen = 100)
```

```r
par(bg = 'white' ,family = "AppleGothic")
plot(doil[,-c(1:3)])
```

```r
dsc <- scale(doil[,4:9])
head(dsc)
```

```r
par( bg='white' , family = "AppleGothic")
plot(
    dsc[,1],
    type = "l",
    col = "red",
    ylim = c(-4, 7),

)

lines(dsc[,3],
    type="l", 
    col = "green",
)
lines(dsc[,5],
    type="l", 
    col = "blue",
)

legend(200, 6, # x, y 좌표값
    c('원달러', '크루드유', '브렌트유'),
    col= c('red', 'blue', 'green'),
    lty = c(1,1,1),
    cex = 0.8, # 크기조절 R에서 기본값 1
    
)
```
