```r
# 2행 2열 차트 그리기 : pch 속성으로 구분해 보기 
par(mfrow = c(2,2) , bg='white')

plot(price, type='o', pch = 5)
plot(price, type='o', pch = 15)
plot(price, type='o', pch = 20 , col='blue')
plot(price, type='o', pch = c(rep(15,8),18,18), 
                    col='orange',
                    cex = c(rep(1,8),3,1),
)
```

# 수학 함수의 시각화 
y = x^2

딥러닝 경사하강법~ 에서 많이 사용한다

```r
f <- function(x){
    y = x^2
    return(y)
}
```

```r
x <- 1:3
f(x)
```

```r
x <- seq(-5, 5, 0.1)

par(bg='white')
plot(x,f(x), 
    type='l'
)
```

y = e^(-x) - 4x^2 + x + 3|x| - 3

```r
f <- function(x){
    y = exp(-x) - 4 * x^2 + x + 3*abs(x) - 3
    return(y)
}
par(bg = 'white')
plot(x, f(x), type = 'l')
```

```r
# y = x^2 + 2x + 3

f <- function(x){
    y = x^2 + 2*x + 3
}
par(bg = 'white')
plot(x, f(x), type = 'l')
# 미분
f <- function(x){
    y = 2*x + 2
}
lines(x, f(x) , type = 'l')
```
