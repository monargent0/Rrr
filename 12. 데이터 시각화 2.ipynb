### 한 화면에 차트 여러개 그려서 비교하기

```r
price <- runif(10,1,100)
par(bg = 'white')
plot(price)
```

```r
# 2행 2열 차트 그리기 : type 속성으로 구분해 보기 
# 같은 셀 안에 있어야 한다. (다른 종류의 그래프도 가능하다)
par(mfrow = c(2,2) , bg='white')

plot(price, type='l')
plot(price, type='o')
plot(price, type='h')
plot(price, type='s')

# barplot(price)
# pie(price)
```

```r
# 2행 2열 차트 그리기 : pch 속성으로 구분해 보기 
par(mfrow = c(2,2) , bg='white')

plot(price, type='o', pch = 5)
plot(price, type='o', pch = 15)
plot(price, type='o', pch = 20 , col='blue')
plot(price, type='o', pch = c(rep(15,8),18,18), 
                    col='orange',
                    cex = c(rep(1,8),3,1),
)
```

## 중첩 자료 시각화 (중요)

```r
x <- c(1,2,3,4,2,4)
y <- rep(2,6)
```

```r
# Method 1) 안씀 - 다른 데이터가 가려짐 
xy.df <- as.data.frame(table(x,y))
xy.df
plot( x, y, 
    pch = 15,
    cex = 2 * xy.df$Freq
)
```

```r
# Method 2) sunflowerplot 을 잘 써야한다 ! 어디 제출할때는 무조건임
x <- c(1,1,1,2,2,2,2,2,2,3,3,4,5,6)
y <- c(1,1,4,2,3,2,2,2,3,2,1,1,1,1)

z <- data.frame(x,y)

par(bg = 'white')
sunflowerplot(z) 
```

```r
# iris의 Sepal Length,Width의 그래프
ir <- data.frame(iris$Sepal.Length,iris$Sepal.Width)

par(bg = 'white')
sunflowerplot(ir) 
```

## chart 2개를 겹쳐서 출력하기 (투명도 조절)

```r
x <- 1:10
mycol = rgb(1,0,0.3, seq(0, 1, 0.1)) # R , G, B, alpha(투명도,0에가까울수록 투명) 

par( bg='white')
dotchart(x ,col='red')
par(new = T) # 현재 차트를 유지하고 그 위에 다른 차트를 생성한다
barplot(x, col = mycol)
```

### 상자 수염 그래프 ( 이상치 확인, )

```r
# 데이터 값이 있는데 이상한 값을 알려줌
abc <- c( 110, 300, 150, 280, 310)
# summary(abc) 를 그래프로 그린것과 비슷하다
# 최대값, 최소값, 중앙값이 표시됨
par(bg = 'white')
boxplot(abc)

par(bg = 'white')
boxplot(abc, col = 'yellow', horizontal = T)
```

```r
def <- c(1,1,1,1,1,2,5)
summary(def)
```

```r
par(bg = 'white')
boxplot(def) # 위에 점 하나 찍힌게 이상치
```
