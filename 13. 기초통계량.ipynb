# 기본 통계량으로 분석해 보기

```r
# 밀가루 사용
flour <- c(3, -2, -1, 0, 1, -2)

# 다이어트 약
diet <- c(-4, 1, -3, -5, -2, -8)

total <- c(flour, diet)
total
```

```r
# 기본 통계량
summary(total)
```

```r
par(bg = 'white',mfrow = c(2,2))
hist(total)

boxplot(total) # 전체 이상치 확인
boxplot(flour, diet) # 각 케이스 확인
```

```r
cor(flour, diet) # 밀가루도 몸무게 감소에 효과가 있는지 확인 
# 상관계수는 순서가 중요하다
```

```r
x <- c(1,9,3)
y <- c(10,20,30)

par(bg = 'white',mfrow = c(2,2))
plot(x, type='l')
plot(y, type='l')

cor(x,y) # 24
# x 데이터 순서 바꿈
x <- c(1,3,9)
plot(x, type='l')
cor(x,y) # 96
# 상관계수 값이 바뀐다. 데이터의 순서가 매우 중요하다@
```

### coffee 데이터셋

```r
cafe <- read.csv("../Data/coffee.csv")
head(cafe)

# 커피와 과일음료를 같이 팔아야 하나? 에 대한 데이터셋
```

```r
# 커피의 빈도수
table(cafe$Coffee) # 커피 판매량의 편차가 크다

# 범위 빈도수 
coffee_Range = seq(0, 60, 10) # 범위
table(cut(cafe$Coffee ,breaks = coffee_Range, right= F) ) #()는 비포함 []는 포함. right = F : (] / [) 를 바꿔준다

# 줄기 - 잎 그림
stem(cafe$Coffee)
```

```r
# 커피와 주스의 변동계수 / 변동계수 cv = 표준편차 / 평균 # 작은값이 빈번하다는뜻이다.
sd(cafe$Coffee) / mean(cafe$Coffee) # 커피의 변동계수
sd(cafe$Fruit.Cup) / mean(cafe$Fruit.Cup) # 과일컵의 변동계수
# 커피판매의 빈도수가 더 높다? 음식? 음식은 회전율이 중요? 공산품은 상관없음

print(sort(cafe$Fruit.Cup)) # 1잔도 안팔린 날이 많다 회전율이 안좋다
print(sort(cafe$Coffee)) # 계속 꾸준히 팔린다
```

```r
par(bg='white')
boxplot(cafe$Coffee,cafe$Fruit.Cup)
```
